Till yet we study about quantitative variables, What if our interest is to study about $Good -or- Bad$, $Like -or- DisLike$, A Categorical outcome like Binary or multiple classess

Let's first introduced with sigmoid activation functions

Formula: $$\sigma(x) = \frac{1}{1+e^{-x}}$$

$$\sigma'(x) = \frac{d}{dx}(\frac{1}{1+e^{-x}})=solve?$$

Also look at the graph of sigmoid function and derivative of sigmoid function

Solve for derivative:
$$\sigma'(x) = \frac{d}{dx}(\frac{1}{1+e^{-x}})$$

$$\implies \frac{d(\frac{1}{1+e^{-x}})}{dx} = \frac{d(1+e^{-x})^{-1}}{d(1+e^{-x})} * \frac{d(1+e^{-x})}{dx}$$

$$\implies-{\frac{1}{(1+e^{-x})^2}} \cdot {-e^{-x}}  = {-(1+e^{-x})^{-2}} \cdot {-e^{-x}} = {(1+e^{-x})^{-2}}\cdot{e^{-x}} = \frac{e^{-x}}{(1+e^{-x})^{-2}}$$


$$\implies \frac{1}{1+e^{-x}} \cdot \frac{(1+e^{-x}-1) }{1+e^{-x}}= \sigma(x)\cdot(1-\sigma(x))$$

$\text {If you look at the graph below output plot how continuous value converted between 0 and 1 using $\sigma(x)$}$

```python
# #Sigmoid Code
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
#  sigmoid functions
def sigmoid(z):
    """sigmoid"""
    return round(1.0 / (1 + np.exp(-z)),2)

#---------
# Derivative of sigmoid functions
def derivative_sigmoid(z):
    '''Derivative of Sigmoid'''
    f_x = round(1.0 / (1 + np.exp(-z)),2)
    return f_x*(1-f_x)

x = np.linspace(-10,10,100) # create even space sequences
sigmoid_op = np.array([sigmoid(i) for i in x]) # get sigmoid values for each x
derivative = np.array([derivative_sigmoid(i) for i in x]) # get sigmoid derivative at each points of x

plt.plot(x, sigmoid_op,'r',label='linspace(-10,10,100)')
plt.plot(x, derivative,'b',label='linspace(-10,10,100)')

plt.axhline(0.5)
plt.axvline(0.0)
# Draw the grid line in background.
plt.grid()

# Title & Subtitle
plt.title('Sigmoid Function')

# place the legen boc in bottom right of the graph
plt.legend(loc='upper left')

# write the Sigmoid formula
plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=15)
plt.text(0.0, 0.3, r"$\sigma'(x)=\sigma(x)\cdot(1-\sigma(x))$", fontsize=15)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

<h3><font color='blue'>Recall linear regression form</font></h3>

$$y = f(x)$$ where $$f(x) = \theta_0 + \theta_1 \cdot x_1$$

Intercept $\theta_0$ & Slope $\theta_1$ are the parameters, If these known you may find the value of $y$

Idea behind classification is to find the class of the given input features

Now, will talk about probability concept called $odd$ ratio

Suppose $p$ be the probability of success, and $1 - p$ be probability of failure than

$$odd(success) = \frac{p}{1-p}$$

Logs of odd will be

$$log(odd) = log(\frac{p}{1-p}) = \theta_i \cdot X_i  \qquad \forall \quad{i}=0,1,2,...,n$$

$$\implies \frac{p}{1-p} = e^{\theta_i \cdot X_i} \implies p = \frac{e^{\theta_i \cdot X_i}}{1+e^{\theta_i \cdot X_i}} \qquad \forall \quad{i}=0,1,2,...,n$$

<font color='darkblue'>__This is known as Loggit Function__</font>

__Note that:__

1.Probability ranges from [0 to 1]

2.Odds range from [0 to $\infty$]

3.Log Odds range from [$-\infty$ to $\infty$]

To get the value between binary range will be using sigmoid function (also known as activation function in deep learning)

$$\sigma(x) = \frac{1}{1+e^{-\theta_i^{T} \cdot X_i}} \qquad \forall \quad{i}=0,1,2,...,n$$

There are various methods to obtain unknown $\theta_i$, Like Maximum Likelihood and Least Squared approach

<font color='blue'>$\text{Logistics Regression is also called Single cell/perceptron neural network which feedforward}$</font>

<font color='blue'>But when we compute gradient of loss function and update the weight, then it's become backward propagation, Term as ANN</font>

<font color='#b86a12'>
Mathematically System of Equation:
$$X_{n \times m} \cdot w_{n \times 1} = b_{m \times 1}$$

Lets take look to loss functions, In regression analysis we studied about Mean Squared Error $\sum{\frac{(\hat{y} - y)^2}{2 \cdot m}}$ or Sum Squared Error $\sum{(\hat{y} - y)^2}$

Logistic Regression we will used Log Loss or Binary Cross Entropy. Entropy is a measure of uncertainty or randomness

Define: Binary Cross Entropy $$ Cost(\sigma(x^{(i)}), y^{(i)}) =J(\theta) = \frac{1}{m} \sum{ - y^{(i)} \cdot log(\sigma(x^{(i)})) - (1-y^{(i)}) \cdot log(1- \sigma(x^{(i)}))}$$

Simple term
 $$J(\theta) = \frac{1}{m} \sum{ - y \cdot log(\sigma(x)) - (1-y) \cdot log(1- \sigma(x))}$$


$$J(\theta) = \begin{cases} {- log(\sigma(x)}& \text{,if } y = 1\\{-log(1- \sigma(x)}& \text{,if } y = 0\end{cases}$$

To fit parameters $\theta$:
$$minimize_\theta -J(\theta)$$

To get prediction on new $x$:
$$\sigma(x) = \frac{1}{1+e^{-\theta^{T} \cdot x}}$$

Note here $\theta_{n \times 1}$ and $X_{n \times m}$ matrices

Compute Gradient Descent of $minimize_{\theta}{ J(\theta)}$ to obtain $\theta$

<font color='purple'>__Repeat__:
    {
    $$\theta_j: \theta_j - \alpha \frac{d}{d\theta_j}J(\theta) \qquad \text{[Simultaneously need to update all weights..]}$$
    }</font>
    
<font color='green'>__Repeat__:
    {
    $$\theta_j: \theta_j - \alpha \cdot \sum_{m=1}^{n} (\sigma_\theta(x^{(i)}) - y^{(i)}) \cdot x^{(i)}_{(j)} \qquad \alpha \text{ is learning rate, How slow or fast want to descent ideal value to choose between 0 & 1}$$
    }</font>
    
<font color='blue'>$$\sum_{m=1}^{n} (\sigma_\theta(x^{(i)}) - y^{(i)}) \cdot x^{(i)}_{(j)} \qquad \text{look like regression algorithm only form has changed instead of}\begin{cases}
\sigma(x)=\theta^T \cdot x & \sigma(x) = Regression\\ 
\sigma(x)=\frac{1}{1+e^{-\theta^T \cdot x}} & \sigma(x) =  classification
\end{cases}$$
</font>

---

<font color = 'purple'>Note: Optimization Derivation

$$\frac{d}{d\theta}J(\theta) = \frac{-1}{m}  \sum{ \frac{d}{d\theta} [y \cdot log(\sigma(x)) + (1-y) \cdot log(1- \sigma(x))]}$$


$$\frac{d}{d\theta}J(\theta) = \frac{-1}{m}  \sum{ \frac{d}{d\theta} [y \cdot log(\frac{1}{1+e^{-\theta \cdot x}}) + (1-y) \cdot log(1- log(\frac{1}{1+e^{-\theta \cdot x}})]}$$

Let's break in steps

Using chain rule will differentiate binary entropy


$$\frac{d J(\theta)}{d\theta} = \frac{d J(\theta)}{dz} \cdot \frac{dz}{dh} \cdot \frac{dh}{d\theta} \qquad \text{where z = $\frac{1}{(1+e^{-h})}$  $\qquad$ h = $\theta^T x$} \qquad equation(*)$$ </font>

<font color='purple'>
    
So
    $$\frac{dJ(\theta)}{dz} = \frac {d ylog(z)} {dz} - \frac {d (1-y)log(1-z)} {dz} \implies  y\frac {d}{dz}log(z) - (1-y)\frac {d}{dz} log(1-z) \implies \frac{y}{z} - \frac{(1-y)}{(1-z)}$$
    
We know sigmoid derivation 
    $$\frac{d}{dh} z = z(1-z)$$ 
    
And 
$$\frac {d}{d\theta} h = x$$ 
</font>

<font color='purple'>
Plugging all piece together in equation(*) we get
$$ - (\frac{y}{z} - \frac{(1-y)}{(1-z)} \cdot z(1-z) \cdot x )=-(\frac{y(1-z)- z(1-y)}{z(1-z)} \cdot z(1-z) \cdot x )$$

Cancelling term we get
$$\frac{d J(\theta)}{d\theta} = -(y-z)\cdot x \qquad \text{remember we had -ve sign in loss function}\\ 
\frac{d J(\theta)}{d\theta} = (z-y)\cdot x $$</font>

<font color='blue'><h3>First Using Gradient Descent Optimization </h3></font>

```python
# Import required methods
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# from sklearn
from sklearn.datasets import load_breast_cancer, load_iris
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

%matplotlib inline
```

```python
# First create all the functions
# Define Activation
def sigmoid(theta, X):
    '''Activation function for binary classification'''
    z = X.dot(theta)
    return 1/(1+np.exp(-z))

#-----------
# Define Loss
def binary_entropy(X , y, theta):
    '''
    caluculate binary cross entropy 
    based on log loss for logistic function
    '''
    p = sigmoid(theta, X)
    m =  len(y)
    J = -np.sum(y * np.log(p) + (1-y) * np.log(1-p))/m
    return J
    
#-----------
# Define Accuracy function
def accuracy(y,pred):
    '''
    y, pred: numpy array
    It will return class and score stack together
    '''
    val = (y == pred)
    val1 = np.unique(val)
    val2 = np.bincount(val)/np.bincount(val).sum(keepdims=True)
    class_score = np.vstack((val1,val2))
    return class_score

#-----------
def classify(predictions):
    '''
    input  - N element array of predictions between 0 and 1
    output - N element array of 0s (False) and 1s (True)
    '''
    decision_boundary = np.vectorize(lambda x: 1 if x>0.5 else 0)
    return decision_boundary(predictions).flatten()
#-----------

# Define Gradient
def gradient(theta, X, y, learning_rate, iterations):
    '''
    Compute gradient descent for theta
    return history of theta, loss and accuracy_score
    '''
    theta_history = [] 
    loss_history = []
    acc_history = []

    for epoch in range(iterations):
        
        # loss
        error = (sigmoid(theta,X)-y)
    
        # gradient
        gradient = X.T.dot(error)
        
        # pred_prob = sigmoid(theta, X)
        y_hat = classify(sigmoid(theta, X))#np.where(pred_prob >= 0.5, 1, 0)
        acc_history.append(accuracy_score(y, y_hat))
        
        # new value of theta
        theta_history.append(theta)
        theta = theta - learning_rate * gradient
        
        # loss history
        loss_history.append(binary_entropy(X, y, theta))
        
    return theta_history , np.array(loss_history), np.array(acc_history)
```

```python
data = load_breast_cancer() # load dataset
minmax = MinMaxScaler() # initialize object
```

```python
X = minmax.fit_transform(data.data) # Scale data usimg minmax between 0-1
y = data.target # target class
```

<font color='purple'><h3>Split Data Into Train & Test</h3></font>

```python
index = np.random.randint(0,len(y), int(len(y)*0.3)) # create index id

Xtrain, Xtest = np.delete(X, index, axis=0) , X[index] # subset based on index id
ytrain, ytest = np.delete(y, index, axis=0) , y[index]
```

```python
print(Xtrain.shape, ytrain.shape) # Verify the shape of data
```

---

```python
rgen = np.random.RandomState(1) # set random state
theta = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1]) # Assign initial theta values
```

```python
# Define hyper parameters for gradient descent
learning_rate= 0.01
itern = 2000
```

```python
%%time
weigths, loss, acc = gradient(theta=theta, X=Xtrain,y=ytrain, learning_rate=learning_rate, iterations=itern)
```

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))

ax[0].plot(range(itern), acc, 'b--')
ax[0].grid()
ax[0].set_title('Model Accuracy')

ax[1].plot(range(itern), loss, 'r.')
ax[1].grid()
ax[1].set_title('Gradient Loss Convergence')
```

```python
idx = np.argmax(acc)
```

```python
weigths[idx]
```

```python
pred_gdo = classify(sigmoid(weigths[idx], Xtest))
```

```python
print(classification_report(ytest, pred_gdo))
```
