<font color='blue'><h2>Basic Statistics:</h2></font>

Mean: $$ \mu = \frac{\sum_{i=1}^n x_i}{n}$$

Median:$$Median =
\begin{cases} 
\frac{X_n}{2} \qquad\qquad \text{if n is even} \\
\frac{X_n + X_n+1}{2} \qquad \text{if n is odd}
\end{cases}$$

Variance:Measure of dispersion from center/mean
$$\sigma^2 = \frac{\sum(x - \mu)^2}{n}$$

Standard Deviation: Square root of variance
$$S = \sqrt\sigma^2 = \sqrt \frac{\sum(x - \mu)^2}{n}$$

```python
from scipy.stats import norm
# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-10, 10, 0.1)
# Mean = 0, SD = 2.
mu = 0
sigma = 3.0
plt.plot(x_axis, norm.pdf(x_axis, mu, sigma), color='black')
plt.axvline(mu, color = 'r')
plt.axvspan(-sigma,sigma, alpha=0.3)
plt.axvspan(-sigma*2,sigma*2, alpha=0.5, color='pink' )
plt.axvspan(-sigma*3,sigma*3, alpha=0.2, color='purple' )
```
