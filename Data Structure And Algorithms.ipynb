```python
def swap(u,v):
    u,v = v,u
    return u,v
    
def bubblesort(lst):
    for j in range(len(lst)):
        for i in range(len(lst)-1):
            if lst[i] > lst[i+1]:
                lst[i],lst[i+1] = swap(lst[i],lst[i+1])
    return lst            

def selectionsort(lst):
    for j in range(len(lst)-1,-1,-1):
        posMax = j
        for loc in range(0,j):
            if lst[loc] > lst[posMax]:
                posMax = loc
        lst[j], lst[posMax] = swap(lst[j], lst[posMax])
    return(lst)
```

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1  or n==2:
        return 1
    else:
        return fibonacci(n-2) + fibonacci(n-1)
```

```python
# Merge sort algorithms
a = [15,18,42,51]
b = [8,11,16,17,44,58,71,74]

def merge(a,b):
    c=[]
    i = 0
    j = 0
    while i <len(a) and j < len(b):
        if a[i] < b[j]:
            c.append(a[i])
            i +=1
        else:
            c.append(b[j])
            j +=1
    while i < len(a):
        c.append(a[i])
        i +=1
    while j < len(b):
        c.append(b[j])
        j +=1
    return c
```

```python
def partition(a,f,l):
    x = a[l]
    i = f-1
    for j in range(f, l+1):
        if a[j]<=x:
            i=i+1
            a[i], a[j] = a[j], a[i]
    return i
        
def quicksort(a, f, l):
    if f < l:
        k = partition(a, f, l)
        quicksort(a, f, k-1)
        quicksort(a, k, l)
```

```python
class Node:
    
    def __init__(self):
        self.data = None
        self.next = None
        
    def setData(self, newdata):
        self.data = newdata
        
    def getData(self):
        return self.data

    def setNext(self, newnext):
        self.next = newnext        
    
    def getNext(self):
        return self.next
    
    def hasNext(self):
        self.next!=None
```

```python
class LinkList():
    
    def __init__(self):
        self.head = None 
    
    def add(self,n):
        temp = Node()
        temp.setData(n)
        temp.setNext(self.head)
        self.head = temp
```

```python
mylist = LinkList()
mylist.add(17)
mylist.add(20)
mylist.add(7)
mylist.add(2)
```

```python
class Stack:
    def __init__(self):
        self.items = []
    def is_empty(self):
        return self.items ==[]
    def push(self, item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def peek(self):
        return self.items[len(self.items)-1]
    def size(self):
        return len(self.items)      

def divBy2(decimal, base=2):
    rstack = Stack()
    while decimal > 0:
        r = decimal %base
        rstack.push(str(r))
        decimal = decimal//base
    return "".join(rstack.items)
```

```python
m = Stack()
m.push('x')
m.push('y')
m.push('z')
print(m.items)
print(m.size())
# print(m.pop())
print(m.peek())
```

```python
class Queue:
    def __init__(self):
        self.items = []
    def is_empty(self):
        return self.items ==[]
    def enqueue(self, item):
        self.items.insert(0,item)
    def dequeue(self):
        return self.items.pop()
    def size(self):
        return len(self.items) 
```

```python
q= Queue()
q.enqueue("hello")
print(q.items)
q.enqueue("dog")
print(q.items)
q.enqueue(3)
print(q.items)
print(q.dequeue())
print(q.items)
```

```python
def Palindrome(item):
    m = Stack()
    r = ''
    for i in item:
        m.push(i)
    while not m.is_empty():
        r = r + str(m.pop())
        print("Check item:",r)
    if r==item:
        print("It's a Palindrome")
    else: 
        print("It's not a Palindrome")
    return r
```
