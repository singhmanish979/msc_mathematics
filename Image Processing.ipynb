```
import pandas as pd
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
import scipy as spy
%matplotlib inline
```

```
# Contours can be explained simply as a curve joining all the continuous points 
# (along the boundary), having same color or intensity. 
# The contours are a useful tool for shape analysis and object detection and recognition.
```

```
image  = io.imread('//content//drive//My Drive//Colab Notebooks//000bc353.jpg')
```

```
io.imshow(image)
```

```
image.shape
```

```
imgray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(imgray, cmap=plt.cm.gray)
```

```
ret,thresh = cv2.threshold(imgray,127,255,0)
```

```
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```
img = cv2.drawContours(imgray, contours, 1, (0,255,0), 3)
plt.imshow(img, cmap=plt.cm.gray)
```

```
from scipy.ndimage.filters import convolve, median_filter, maximum_filter, minimum_filter, sobel, prewitt
import skimage
```

```
a = skimage.color.rgb2gray(image)
plt.imshow(a, cmap=plt.cm.gray)
```

```
# initializing the filter of size 5 by 5
# the filter is divided by 25 for normalization
k = np.ones((5,5))/25
# performing convolution
b = convolve(a, k)
print(b.shape)
plt.imshow(b, cmap=plt.cm.gray)
```

```
b =  median_filter(a,size=5)
print(b.shape)
plt.imshow(b, cmap=plt.cm.gray)
```

```
b =  maximum_filter(a,size=5)
print(b.shape)
plt.imshow(b, cmap=plt.cm.gray)
```

```
b =  minimum_filter(a,size=5)
print(b.shape)
plt.imshow(b, cmap=plt.cm.gray)
```

```
b =  sobel(a)
print(b.shape)
plt.imshow(b,cmap=plt.cm.gray)
```

```
b =  prewitt(a)
print(b.shape)
plt.imshow(b,cmap=plt.cm.gray)
```

```
from skimage.filters import sobel, sobel_h, sobel_v, prewitt, prewitt_v, prewitt_h, laplace
from skimage.feature import canny
```

```
plt.imshow(a, cmap=plt.cm.gray)
```

```
b =  canny(a, sigma=4)
print(b.shape)
plt.imshow(b,cmap=plt.cm.gray)
```

```
c = median_filter(b, size=2)
print(c.shape)
plt.imshow(c,cmap=plt.cm.gray)
```

```
from scipy.ndimage.filters import laplace, gaussian_laplace
```

```
b = laplace(a)
print(b.shape)
plt.imshow(b,cmap=plt.cm.gray)
```

```
b = gaussian_laplace(a, sigma=3.0)
print(b.shape)
plt.imshow(b,cmap=plt.cm.gray)
```
